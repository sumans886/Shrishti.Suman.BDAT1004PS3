```python
#2. the Male ratio per occupation and sort it from the most to the least

genderTotal = users['gender'].value_counts()
genderOccupation = users.groupby('gender')['occupation'].value_counts()
sexRatio = genderOccupation['M']/genderOccupation['F']
print(sexRatio.sort_values(ascending=False))
```

```python
#3. max and min age by occupation
maxAge =  userOccupation["age"].groupby('occupation').max()
minAge =  userOccupation["age"].groupby('occupation').min()
print(maxAge, minAge)
```

```python
#4. For each combination of occupation and sex, calculate the mean age
users.groupby(['occupation','gender'])['age'].mean()
```

```python
#5. For each occupation present the percentage of women and men
genderTotal = users['gender'].value_counts()
genderOccupation = users.groupby('gender')['occupation'].value_counts()
maleSexRatio = genderOccupation['M']/(genderOccupation['M'] + genderOccupation['F'])*100
femaleSexRatio = genderOccupation['F']/(genderOccupation['M'] + genderOccupation['F'])*100
print(maleSexRatio, femaleSexRatio)
```

Assign it to a variable called data and replace the first 3 columns by a proper datetime index. Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
```python
import pandas as pd
import datetime

url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data"
data = pd.read_csv(url, sep="\s+", parse_dates=[[0,1,2]])

data
```

```python
#  Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
def fixDate(d):
    year = d.year - 100 if d.year > 1989 else d.year
    return datetime.date(year, d.month, d.day)
    
data['Yr_Mo_Dy']  = data['Yr_Mo_Dy'].apply(fixDate)

data
```

```python
#  Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
data = data.set_index('Yr_Mo_Dy')
data
```

```python
data.index
```

```python
#  Number of non-missing values for each location
data.isnull().sum()
```

```python
# Compute how many non-missing values there are in total.
data.notnull().sum().sum()
```

```python
# Calculate the mean windspeeds of the windspeeds over all the locations and all the times
data.sum().sum() / data.notna().sum().sum()
```

```python
data.describe(percentiles=[])
```

```python
# Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
dayStats = pd.DataFrame()

dayStats['min'] = data.min(axis = 1) # min
dayStats['max'] = data.max(axis = 1) # max 
dayStats['mean'] = data.mean(axis = 1) # mean
dayStats['std'] = data.std(axis = 1) # standard deviations

dayStats.head()
```

```python
# Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day. A different set of numbers for each day.
dayStats = pd.concat([data.min(axis=1), data.max(axis=1), data.mean(axis=1), data.std(axis=1)], axis=1)

# Rename the columns with proper labels
dayStats.rename(columns={0:'Min',1:'Max',2:'Mean',3:'Std'}, inplace=True)

dayStats
```

```python
# Find the average windspeed in January for each location.
data.loc[data.index.month == 1].mean()
```

```python
# Downsample the record to a yearly frequency for each location
data.groupby(data.index.to_period('A')).mean()
```

```python
# Downsample the record to a monthly frequency for each location.
data.groupby(data.index.to_period('M')).mean()
```

```python
weekly = data.resample('W').ffill()

weekly
```

```python
# Downsample the record to a weekly frequency for each location.

data.groupby(data.index.to_period('W')).mean()
```

```python
# Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.
weekly_stats = weekly[:52]

weekly_stats.shape[0] # Total 52 weeks in 1961
```

```python
weekly_stats = pd.concat([weekly_stats.mean(), weekly_stats.max(), weekly_stats.mean(), weekly_stats.std()], axis=1)
weekly_stats.rename(columns={0:'Mean',1:'Max',2:'Mean',3:'Std'},inplace=True)

weekly_stats
```

```python
# What was the most ordered item in the choice_description column?
mostInChoiceDesc = chipo.groupby(['choice_description']).sum().sort_values('quantity', ascending=False).head(1)

mostInChoiceDesc
```

```python
# How many items were orderd in total?
print(f'Items ordered in total = {chipo.quantity.sum()}')
```

```python
temp_df = chipo['item_price'].str.split('$', n = 1, expand = True)
  
chipo['temp_price']= temp_df[1]

# Since, we have extracted the price in 'temp_df', let's delete the existing 'item_price' column
chipo.drop(columns =['item_price'], inplace = True)

# rename 'temp_price' column back to 'item_price'
chipo.rename(columns = {'temp_price':'item_price'}, inplace = True)

# Let's set the datatype of 'item_price' column to 'float'
chipo.item_price.astype(float)
```

```python
# Check the item price type
chipo.item_price.dtype # Check the datatype of 'item_price' column
```

```python
# Create a lambda function and change the type of item price
chipo.item_price = chipo.item_price.apply(lambda x : float(x[1:-1]))
```

```python
# Check the item price type
chipo.item_price.dtype
```

```python
# How much was the revenue for the period in the dataset?
revenue = (chipo.quantity * chipo.item_price).sum()

print(f'Total revenue for the given period is ${revenue}')
```

```python
# How many orders were made in the period?
orders = chipo.order_id.value_counts().count() # order_id has repetitive data while we need unique orders only 

print(f'Total number of orders in the given period are {orders}')
```

```python
# What is the average revenue amount per order?
tempChipo = chipo.loc[:,('order_id', 'quantity', 'item_price')]

tempChipo['revenue'] = tempChipo['quantity'] * tempChipo['item_price'] 

avg = tempChipo.groupby(by=['order_id']).sum().mean()['revenue']

print(f'Average revenue amount per order is {avg}')
```

```python
# How many different items are sold?
items_sold = chipo.item_name.value_counts().count()

print(f'Total different items sold = {items_sold}')
```

# Question 7 
Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000. Don't forget to label your axes!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("us-marriages-divorces-1867-2014.csv")

# data
marriages_per_capita = data['Marriages_per_1000']
divorces_per_capita = data['Divorces_per_1000']

year = [1900, 1950, 2000]
ind = ["marriages per capita", "divorces per capita"]

specific_data = data.loc[data['Year'] == 1900]
specific_data = specific_data.append(data.loc[data['Year'] == 1950])
specific_data = specific_data.append(data.loc[data['Year'] == 2000])
columns = specific_data['Marriages_per_1000'], specific_data['Divorces_per_1000']
plt.style.use('dark_background')
fig, ax = plt.subplots()
bar_width = 0.35
X = np.arange(3)

ax = plt.bar(X, columns[0], bar_width, label="Number of Marriages")
ax = plt.bar(X + bar_width, columns[1], bar_width, label="Number of Divorces")

plt.xlabel('Year')
plt.ylabel('Marriage/Divorces per 1000 people')
plt.xticks(X + (bar_width/2) , ("1900", "1950", "2000"))
plt.legend()
plt.tight_layout()
```
